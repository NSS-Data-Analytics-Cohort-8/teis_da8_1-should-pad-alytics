```python
colors = ["#6E99EC", "#777776", "#EF412A"]

# Increase the figure size
plt.figure(figsize=(12, 8))

# Plot the bar chart with adjusted bar width and edgecolor
sns.barplot(x=total_count['IFSP Teaming ID name'], y=total_count['Total Count'], palette=colors, linewidth=1.5, edgecolor='black')

plt.title('Count of IFSP Teaming IDs for Telehealth Services', fontsize=16)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Remove x and y axis labels
plt.xlabel('')
plt.ylabel('')

# Add data labels to the bars
for i, count in enumerate(total_count['Total Count']):
    plt.text(i, count + 1, str(count), ha='center', va='bottom', fontsize=10)

# Add a grid
plt.grid(True, axis='y', linestyle='--')

# Remove the top and right spines
sns.despine()

plt.tight_layout()
plt.show()
```
