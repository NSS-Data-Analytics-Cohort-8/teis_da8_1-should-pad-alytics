```python
df_telehealth = df1[df1['SERVICE_SETTING'] == 'Telehealth']
df3 = df_telehealth.groupby(['CHILD_ID', 'SERVICE_DATE', 'START_TIME', 'IFSP Teaming ID name']).size().reset_index(name='Count')
total_count = df3.groupby(['IFSP Teaming ID name']).size().reset_index(name='Total Count')

# Calculate the total count of all three combined
total_all_combined = total_count['Total Count'].sum()

# Create a new row with the total count of all three combined
total_combined_row = pd.DataFrame({'IFSP Teaming ID name': ['Total'], 'Total Count': [total_all_combined]})
total_count = pd.concat([total_count, total_combined_row], ignore_index=True)
total_count
```

```python
# Sort in descending order
total_count_sorted = total_count.sort_values('Total Count', ascending=False)

colors = ["#6E99EC", "#777776", "#EF412A", "#C7CAD5"]

# Increase the figure size
plt.figure(figsize=(12, 8))

# Plot the bar chart with adjusted bar width and edgecolor
sns.barplot(x=total_count_sorted['IFSP Teaming ID name'], y=total_count_sorted['Total Count'], palette=colors, linewidth=1.5, edgecolor='black')

plt.title('Count of IFSP Teaming IDs for Telehealth Services', fontsize=16)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Remove labels
plt.xlabel('')
plt.ylabel('')

# Add data labels to the bars
for i, count in enumerate(total_count_sorted['Total Count']):
    plt.text(i, count + 1, str(count), ha='center', va='bottom', fontsize=10)

# Add grid
plt.grid(True, axis='y', linestyle='--')

# Remove the top and right spines
sns.despine()

plt.tight_layout()
plt.show()
```
