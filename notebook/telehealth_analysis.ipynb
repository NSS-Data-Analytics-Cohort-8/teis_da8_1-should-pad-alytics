5. What is the rate of telehealth that are Co visits versus other services? Is telehealth used for a specific type of FSP Teaming ID Name?


```python
df1 = pd.read_excel('../data/TEIS_IFSP_Teaming_Jan_2022-_Jan_2023.xlsx', sheet_name='Compiled_IFSP Teaming')
df1
```

```python
df2 = pd.read_excel('../data/TEIS_IFSP_Teaming_Jan_2022-_Jan_2023.xlsx', sheet_name='All Children with IFSPs')
```

```python
co_visits = df1[df1["IFSP Teaming ID name"] == "Co-Visit"]
co_visits
```

```python
telehealth_co_visits = co_visits[co_visits["SERVICE_SETTING"] == "Telehealth"]
telehealth_co_visits
```

```python
telehealth_co_visit_count = len(telehealth_co_visits)
telehealth_co_visit_count
```

```python
telehealth_co_visit_frequency = telehealth_co_visits["SERVICE_DATE"].value_counts()
telehealth_co_visit_frequency
```

```python

telehealth_co_visits['SERVICE_DATE'] = pd.to_datetime(telehealth_co_visits['SERVICE_DATE'])

co_visits_by_month = telehealth_co_visits.groupby(telehealth_co_visits['SERVICE_DATE'].dt.to_period('M')).size()

co_visits_by_month.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Telehealth Co-Visits by Month')
plt.xlabel('Month')
plt.ylabel('Count of Co-Visits')
plt.grid(True)
plt.show()
```

```python
telehealth_co_visits['START_TIME'] = pd.to_timedelta(telehealth_co_visits['START_TIME'])
telehealth_co_visits['END_TIME'] = pd.to_timedelta(telehealth_co_visits['END_TIME'])

telehealth_co_visits['START_DATETIME'] = telehealth_co_visits['SERVICE_DATE'] + telehealth_co_visits['START_TIME']
telehealth_co_visits['END_DATETIME'] = telehealth_co_visits['SERVICE_DATE'] + telehealth_co_visits['END_TIME']

telehealth_co_visits['DURATION'] = telehealth_co_visits['END_DATETIME'] - telehealth_co_visits['START_DATETIME']

# Calculate summary statistics of the duration column
duration_stats = telehealth_co_visits['DURATION'].describe()
print(duration_stats)
```

count: The number of non-null values in the DURATION column, which is 775 in this case.

mean: The average duration of the telehealth co-visits. The mean value is 0 days 00:55:14.012903225, indicating that, on average, the duration of a co-visit is approximately 55 minutes and 14 seconds.

std: The standard deviation of the duration values. The standard deviation is 0 days 00:18:31.932589018, suggesting that the durations tend to deviate from the mean by approximately 18 minutes and 31 seconds.

min: The minimum duration value in the DURATION column is 0 days 00:15:00, indicating that the shortest co-visit duration recorded is 15 minutes.

25%: The first quartile of the duration values. The value 0 days 00:45:00 means that 25% of the co-visits had a duration of 45 minutes or less.

50%: The median of the duration values. The median, also known as the second quartile, is 0 days 01:00:00, implying that 50% of the co-visits lasted for 1 hour or less.

75%: The third quartile of the duration values. The value 0 days 01:00:00 indicates that 75% of the co-visits had a duration of 1 hour or less.

max: The maximum duration value in the DURATION column is 0 days 03:00:00, meaning that the longest co-visit duration recorded is 3 hours

```python
top_10_agencies = telehealth_co_visits['Agency Name'].value_counts().nlargest(10)

plt.figure(figsize=(10, 6))
top_10_agencies.plot(kind='bar')
plt.xlabel('Agency')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout() 
plt.show()
```

```python
lowest_10_agencies = telehealth_co_visits['Agency Name'].value_counts().nsmallest(10)

plt.figure(figsize=(10, 6))
lowest_10_agencies.plot(kind='bar')
plt.xlabel('Agency')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout() 
plt.show()
```

```python
user_counts = telehealth_co_visits['ADD_USER'].value_counts().head(10)

plt.figure(figsize=(10, 6))
user_counts.plot(kind='bar')
plt.title('Count of Services Delivered by Top 10 Users')
plt.xlabel('User')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
telehealth_co_visits['DAY_OF_WEEK'] = telehealth_co_visits['START_DATETIME'].dt.day_name()

day_of_week_counts = telehealth_co_visits['DAY_OF_WEEK'].value_counts()

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
day_of_week_counts.plot(kind='bar')
plt.title('Co-Visits by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
telehealth_co_visits['HOUR_OF_DAY'] = telehealth_co_visits['START_DATETIME'].dt.hour

hour_of_day_counts = telehealth_co_visits['HOUR_OF_DAY'].value_counts()

hour_of_day_counts = hour_of_day_counts.sort_index()

plt.figure(figsize=(10, 6))
hour_of_day_counts.plot(kind='bar')
plt.title('Co-Visits by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Filter the data for the specified IFSP Teaming ID name
covisit_data = df1[df1['IFSP Teaming ID name'] == 'Co-Visit']

# Count the number of telehealth visits
telehealth_count = covisit_data[covisit_data['SERVICE_SETTING'] == 'Telehealth'].shape[0]

# Count the number of other service settings
other_count = covisit_data[covisit_data['SERVICE_SETTING'] != 'Telehealth'].shape[0]

print("Telehealth Count:", telehealth_count)
print("Other Service Settings Count:", other_count)
```

```python
# Define the data
categories = ['Telehealth', 'Other Service Settings']
counts = [telehealth_count, other_count]

# Create the bar plot
plt.bar(categories, counts)

# Add labels and title
plt.xlabel('Service Setting')
plt.ylabel('Count')
plt.title('Count of Telehealth Visits vs Other Service Settings during covisits')

# Display the plot
plt.show()
```

```python

# Filter the data for telehealth visits
telehealth_data = df1[df1['SERVICE_SETTING'] == 'Telehealth']

# Count the number of telehealth visits for each FSP Teaming ID Name
telehealth_counts = telehealth_data['IFSP Teaming ID name'].value_counts()

# Create the bar plot
telehealth_counts.plot(kind='bar')

# Add labels and title
plt.xlabel('FSP Teaming ID Name')
plt.ylabel('Count of Telehealth Visits')
plt.title('Count of Telehealth Visits by FSP Teaming ID Name')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Filter the data for co-visits
co_visit_data = df1[df1['IFSP Teaming ID name'] == 'Co-Visit']

# Count the number of co-visits that are telehealth
telehealth_count = co_visit_data[co_visit_data['SERVICE_SETTING'] == 'Telehealth'].shape[0]

# Count the number of co-visits that are not telehealth
other_count = co_visit_data[co_visit_data['SERVICE_SETTING'] != 'Telehealth'].shape[0]

# Create the pie chart
labels = ['Telehealth', 'Other']
sizes = [telehealth_count, other_count]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Add title
plt.title('Rate of Co-Visits: Telehealth vs Other')

# Set aspect ratio to be equal, so the pie is circular
plt.axis('equal')

# Display the pie chart
plt.show()
```
